### Data Cleaning

```python
#Here, we drop ID and year since these variables aren't necessary for analysis
df.drop(['ID','year'],axis=1, inplace=True)
```

```python
#Check for missing values
df.isnull().sum()
```

```python
#Remove columns with thousands of missing values 
df.drop(['LTV','dtir1', 'Interest_rate_spread', 'rate_of_interest', 'Upfront_charges', 'property_value'],axis=1, inplace=True)
```

```python
df.shape
```

```python
#Since this dataset contains 148k rows, we remove the rest of the rows with missing values
df.dropna(subset=["loan_limit", "approv_in_adv", "loan_purpose", "term", "Neg_ammortization", "income", "age", "submission_of_application"], axis=0, inplace=True)
```

```python
df.shape
```

```python
df.to_csv("../Data/cleaned_loan.csv")
```

```python
#Create variable for categorical data and numerical data to prep for visualization 
default_status = df['Status']
X_data = df.drop(['Status'], axis = 1)
features = list(X_data)
unique_n = X_data.nunique().sort_values()
display(unique_n)
```

All categorical variables have 7 or less unique values. 

```python
numerical_col = []
categorical_col = []

for i in range(len(X_data.columns)):
    if unique_n.values[i]<=7:
        categorical_col.append(unique_n.index[i])
    else: 
        numerical_col.append(unique_n.index[i])
        
print("There are", len(numerical_col), "numerical columns")
print("There are", len(categorical_col), "categorical columns")
```
